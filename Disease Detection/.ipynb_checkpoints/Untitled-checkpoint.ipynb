```python
import pandas as pd
import numpy as np
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,MaxPool2D,Dropout
from sklearn.preprocessing import LabelEncoder
from tensorflow import keras
from tensorflow.keras.regularizers import l2
```

```python
Eyes=pd.read_csv(r"C:\Users\rishi\Desktop\Eyes_Disease.csv")
Brain=pd.read_csv(r"C:\Users\rishi\Desktop\Brain_Disease.csv")
Heart=pd.read_csv(r"C:\Users\rishi\Desktop\Heart_Disease.csv")
Kidney=pd.read_csv(r"C:\Users\rishi\Desktop\Kidney_Disease.csv")
```

```python
Brain.colums=Brain.columns.str.strip()
Brain[Brain.columns[0]] = Brain[Brain.columns[0]].str.strip()
Brain.drop_duplicates(subset=Brain.columns[0],inplace=True)
```

```python
Eyes.colums=Eyes.columns.str.strip()
Eyes[Eyes.columns[0]]=Eyes[Eyes.columns[0]].str.strip()
Eyes.drop_duplicates(subset=Eyes.columns[0],inplace=True)
```

```python
Heart.colums=Heart.columns.str.strip()
Heart[Heart.columns[0]]=Heart[Heart.columns[0]].str.strip()
Heart.drop_duplicates(subset=Heart.columns[0],inplace=True)
```

```python
Kidney.colums=Kidney.columns.str.strip()
Kidney[Kidney.columns[0]]=Kidney[Kidney.columns[0]].str.strip()
Kidney.drop_duplicates(subset=Kidney.columns[0],inplace=True)
```

```python
def Model(data):
    X=data.iloc[:,1:]
    Y=data.iloc[:,0]
    label_encoder = LabelEncoder()
    Y_encoded = label_encoder.fit_transform(Y)
    name=Sequential()
    name.add(Dense(512,activation='relu',input_shape=(X.shape[1],),kernel_regularizer=l2(0.01)))
    name.add(Dense(128,activation='relu'))
    name.add(Dense(64,activation='relu'))
    name.add(Dense(32,activation='relu',kernel_regularizer=l2(0.01)))
    name.add(Dense(128,activation='relu'))
    name.add(Dense(64,activation='relu',kernel_regularizer=l2(0.01)))
    name.add(Dense(32,activation='relu'))
    name.add(Dense(len(label_encoder.classes_),activation='softmax'))
        
    name.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
    history=name.fit(X,Y_encoded,epochs=500,validation_data=(X,Y_encoded))
    return name
```

```python
Model_Brain=Model(Brain)
```

```python
Model_Eyes=Model(Kidney)
```

```python
Model_Heart=Model(Heart)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
def Model_DesionTree(data):
    X=data.iloc[:,1:].values
    Y=data.iloc[:,0]
    
    clf = RandomForestClassifier(n_estimators=200,random_state=42)

    # Train the classifier on the training data
    clf.fit(X, Y)

    # Make predictions on the test data
    Y_pred = clf.predict(X)

    # Evaluate the model's performance
    accuracy = accuracy_score(Y, Y_pred)
    print(f"Accuracy: {accuracy:.2f}")

    # Display a classification report
    print(classification_report(Y, Y_pred))
    return clf
```

```python
Kidney_Model=Model_DesionTree(Kidney)
```

```python
import joblib

# Assuming 'clf' is your trained SVC classifier
model_filename = 'Kidney.pkl'
joblib.dump(Kidney_Model, model_filename)
```
